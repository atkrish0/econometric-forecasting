```python
import pandas as pd
import numpy as np
```

### Import data

```python
file_path = 'CMO-Historical-Data-Monthly.xlsx'
Prices = pd.read_excel(file_path, sheet_name= 'Monthly Prices', skiprows=4)
```

```python
Prices
```

### Process raw commodity data

```python
# Convert the 'Date' column to datetime, specifying the correct format
Prices['Date'] = pd.to_datetime(Prices['Date'], format='%YM%m')

# Set the 'Date' column as the index to create a timetable-like structure
Prices = Prices.set_index('Date')
```

### Creating commodity classes

```python
# Get the total number of columns
total_columns = len(Prices.columns)

# Iterate through each column to display its name and number of rows
print(f"Total number of columns: {total_columns}\n")
i=1
for column_name in Prices.columns:
    num_rows = Prices[column_name].count()
    print(i)
    print(f"Column Name: {column_name}")
    print(f"Number of Rows: {num_rows}")
    print('-' * 40)  # Separator for readability
    i+=1
```

```python
# Define the number of commodities in each category
categories = {
    "Energy": 10,
    "Beverages": 7,
    "Oils and Meals": 11,
    "Grains": 9,
    "Other Food": 10,
    "Raw Materials": 9,
    "Fertilizers": 5,
    "Metals and Minerals": 7,
    "Precious Metals": 3
}

# Create the commodity classes list
commodity_classes = []
for category, count in categories.items():
    commodity_classes.extend([category] * count)

# Convert the list to a pandas Categorical type
commodity_classes = pd.Categorical(commodity_classes)

# Assign the commodity classes, ensuring the first commodity is classified
Prices.columns = [f"{name} ({commodity_classes[i]})" if i < len(commodity_classes) else name for i, name in enumerate(Prices.columns)]
```

```python
# Print all column names to verify the classification
for column in Prices.columns:
    print(column)
```

```python
commodity_classes
```

```python
# Step 1: Initialize all commodities as "Soft"
commodity_types = pd.Series(["Soft"] * len(commodity_classes), dtype="category")

# Step 2: Define possible categories ("Soft" and "Hard")
commodity_types = commodity_types.cat.add_categories(["Hard"])

# Step 3: Define which classes are considered "Hard"
hard_classes = ["Energy", "Fertilizers", "Metals and Minerals", "Precious Metals"]

# Step 4: Identify the indexes of "Hard" commodities
hard_idx = commodity_classes.isin(hard_classes)

# Step 5: Update the commodity types for "Hard" commodities
commodity_types[hard_idx] = "Hard"

# Step 6: Create a DataFrame for Commodity Types, associating them with the column names
commodity_types_df = pd.DataFrame({
    "Commodity": Prices.columns[1:len(commodity_classes) + 1],  # Skip the date column
    "CommodityType": commodity_types.values
})

# Print the classification to verify
print(commodity_types_df)
```
